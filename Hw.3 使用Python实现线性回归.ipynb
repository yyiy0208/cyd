```python
import pandas as pd
import numpy as np

class LinearRegression:

    def __init__(self, tol=0.0002):
        self.tol = tol

    def fit(self, X, y):
        X['constant'] = 1
        n_features = X.shape[1]
        stat_check = 1

        beta = np.array([0.0] * n_features)
        epsilon = y
        for j in range(n_features):
            epsilon = epsilon - beta[j] * X.iloc[:, j]

        while np.mean(epsilon) > self.tol:
            
            if np.mean(epsilon) > 0.001:
                alpha = 0.1
            elif np.mean(epsilon)>0.0001:
                alpha = 0.01
            else:
                alpha = 0.001
            for j in range(n_features):
                gradient = -np.mean(epsilon * X.iloc[:, j])
                beta[j] = beta[j] - alpha * gradient
            epsilon = y
            for j in range(n_features):
                epsilon = epsilon - beta[j] * X.iloc[:, j]
               
            
        self.coef_ = beta[:-1]
        self.intercept_ = beta[-1]
        return self

    def predict(self, X):
        return np.dot(X, self.coef_) + self.intercept_
```
